```python
def login(user_type,username,password):
    """
Function to authenticate username and password.
"""
    sql = "SELECT user_type,id,password FROM user_pass WHERE id = %s AND user_type = %s"
    val = (username,user_type)
    mycursor.execute(sql,val)

    results = mycursor.fetchall()
    #print(results)

    if len(results) == 0:
        print("No such user present in database")
        return False,None
    
    if len(results) == 1:
        #hashed_pass = hashlib.pbkdf2_hmac('sha256', password.encode('utf-8'), salt, 100000)
        
        if results[0][2] == password:
            
            ts = time.time()
            
            timestamp = datetime.fromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
            '''print(timestamp)
            print(results[0][0])
            print(results[0][1])
            print(results[0][2])'''
        

            sql = "INSERT INTO log_details(user_id,login_time) VALUES (%s,%s)"
            val = (results[0][1],timestamp)

            mycursor.execute(sql,val)
            mydb.commit()
            #print(mycursor.fetchall())
            print("Login successful")
            time.sleep(1)
            
            return True,results[0][1]
        else:
            print("Login Denied: Wrong password")
            return False,None
    else:
        print("Multiple users present. Contact Admin")
        return False,None
```

```python
def logout(user_id):
    sql = "DELETE FROM log_details WHERE user_id =  %s"
    val = (user_id,)

    mycursor.execute(sql,val)
    mydb.commit()
    print("logged out successfully:",user_id)
    
```
